```
"""
Created on Sat Aug 31 13:41:34 2019
Program: Decision Tree Regression
Author: Pramodkumar Gupta
"""

# Import all libraries
import matplotlib.pyplot as plt
import pandas as pd

# Read data into pandas dataframes

df=pd.read_csv('Position_Salaries.csv')
```

```
# Slice the data in indepedent and depedent variables

X=df.iloc[:,1:2].values              
y=df.iloc[:,2].values 
```

```
import seaborn as sns
sns.pairplot(df,hue="Salary")
```

```
from sklearn.tree import DecisionTreeRegressor  
model = DecisionTreeRegressor(random_state = 0)  
model.fit(X, y) 

y_pred = model.predict(X)
```

```
# Plotting the graph
plt.scatter(X,y,color='blue')
plt.plot(X,y_pred,color='green' )           # Green - Decision Tree Regression
plt.title('Expireince Salary Graph')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.figure(figsize=(12,7))
plt.show()
```
