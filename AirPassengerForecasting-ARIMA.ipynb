#### This is for doing Air Passenger Forecasting using ARIMA time series Model

```python
# Import All libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10,6
```

```python
# Load data into pandas dataframe

df=pd.read_csv('Dataset/Air Passenger.csv')
```

```python
df['Month']=pd.to_datetime(df['Month'],infer_datetime_format=True)
df_index=df.set_index(['Month'],inplace=True)
```

```python
# Visualize the data
plt.plot(df)
plt.title('Time Series Data Pattern')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
```

```python
df_log=np.log(df)
plt.plot(df_log)
```

```python
df=np.exp(df_log)
plt.plot(df)
```
